# 3. 파생변수 생성

```python
## 출루율
private_batter.columns
private_batter["OBP"]=(private_batter["HIT"]+private_batter["BB"]+private_batter["HP"]+private_batter["IB"])/private_batter["PA"]
private_batter["OBP"].fillna(0,inplace=True)
private_batter[["HIT","H2","H3","HR","PA","OBP"]]
```

```python
## 장타율
private_batter["SLG"]=(private_batter["HIT"] + 2*private_batter["H2"] + 3*private_batter["H3"] + 4*private_batter["HR"])/private_batter["AB"]
private_batter['SLG'].fillna(0, inplace=True)
private_batter[["HIT","H2","H3","HR","AB","SLG"]]
```

```python
## OPS = OBP(출루율) + SLG(장타율)

private_batter["OPS"]=private_batter["OBP"]+private_batter["SLG"]
private_batter[["OBP","SLG","OPS"]]
```

```python
## GPA = (1.8*출루율 + 장타력)/4

private_batter["GPA"]=(1.8*private_batter["OBP"] + private_batter["SLG"])/4
private_batter[["OBP","SLG","GPA"]]
```

```python
## SB_trial (도루 시도)
private_batter['SB_trial'] =  private_batter['SB'] + private_batter['CS']

## SB_SR (도루 성공율)
private_batter['SB_SR'] =  private_batter['SB'] /(private_batter['SB'] + private_batter['CS'])
private_batter['SB_SR'].fillna(0, inplace=True)
private_batter[["SB","CS","SB_trial","SB_SR"]]
```
