# 문자로 되어 있는 것을 labelencoding

```python
from sklearn.preprocessing import LabelEncoder
cat_features = ['CARD_SIDO_NM', 'STD_CLSS_NM', 'HOM_SIDO_NM','AGE'] # 말 그대로 
# here, don't apply FLC!
df_num[cat_features]= df_num[cat_features].apply(LabelEncoder().fit_transform)
```

```python
df_num.head()
```

```python
real_cat_features =  ['SEX_CTGO_CD','CARD_SIDO_NM', 'STD_CLSS_NM', 'HOM_SIDO_NM','AGE','FLC']
for i in enumerate(real_cat_features):
    ca = i[1]
    df_num[ca] = df_num[ca].astype('category')
```

```python
df_num.dtypes
```

# NA 값 없음 확인

```python
df_num.isna().sum()
```

```python
train_num = df_num.sample(frac=1, random_state=2020)
train_features = train_num.drop(['AMT'], axis=1)  #1114451, 15
train_target = np.log1p(train_num['AMT']) #1114451,

train_features
```

```python
train_target
```

```python
x = train_features
y = train_target


k = int(len(x)*0.7) # 70% chosen

x_train = x[:k]
y_train = y[:k]
x_val = x[k:]
y_val = y[k:]
```

```python
x_train.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in x_train.columns]
x_val.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in x_val.columns]
```

# Light GBM, Adaboostregression, RFregression 조합으로 할 때는 문제가 없었으며
