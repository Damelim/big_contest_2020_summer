## 도움이 될 만한 것들?

- WAR(대체선수 대비 승리기여도)
   => 높을 수록 잘하는 선수
   
   => 팀 타자/투수 별 합산 WAR가 높으면 잘하는 선수들이 많다는 것이므로 팀 타율이 높거나 팀 평균 자책점이 낮지 않을지?
   
   => 타자는 경기에 매일 나오므로 투수에 비해 높게 측정된다는 것이 단점일 수 있으나 타자는 타자끼리, 투수는 투수끼리 비교한다면    
     이는 큰 문제가 되지 않을 것이라 생각
   

- 야구선수의 전성기는 대체적으로 20대 후반~30대 초반
   => 반영할 수 있다면 좋을 수도...
   
   
- 파크팩터: 이 구장이 타자에게 유리한지 투수에게 유리한지 판별하는 수치. 이 값이 클 수록 타자에게 유리 (1인 경우 중립적)
  
   ==> 타자의 성적에 '구장 빨'을 받았는지 알아보기 위해 만들어진 지표
   
   ==> 파크팩터 값이 낮은 구장을 홈구장으로 사용하는 구단의 타자들의 ops는 (+) 보정을 해줘야 할 것으로 보임

## <팀 승률 예측>

### - 당연하게도 점수를 많이 내고, 적게 주는 팀이 승률이 높을 것

=> 팀 타율과 (+)의 관계, 팀 방어율과 (-)의 관계가 있겠지만, 팀 타율과 팀 방어율도 예측해야 함 (불확실) => 다른 대안 필요

=> 각 팀의 타자와 투수의 WAR 합산을 바탕으로 순위를 매겨본 결과 현재 순위와 비슷함 (출처: kbreport.com)    
       그러나 순위 예측이 아닌 승률 예측이 목표이므로 이것을 그대로 사용하기에는 다소 무리가 있을 듯     

### - 그렇다면 팀 별 득점과 실점을 예측해서 이것을 승률로 변환해보면 어떨지?

### 피타고리안 승률

=> 득점, 실점을 바탕으로 팀의 기대승률을 계산하는 방법.

=> 승률 = 득점^2 / (득점^2 + 실점^2)

=> 단순한 수식이지만 꽤 정확하다.

=> 증명: http://suxism.com/?p=3973



### - 어느 팀이 점수를 많이 낼 것인지 어떻게 예측?

1. 팀별 OPS 비교   (*OPS = 출루율 + 장타율)
   : OPS가 높을 수록 (득점)생산성이 높다는 뜻. 득점력을 매우 잘 설명해줌
     따라서 팀 OPS가 높으면 팀 득점 또한 높을 것이다.
   - 단점
   
     1) 출루율과 장타율의 scale이 다름에도 단순 합산한 것이라 출루율에 비해 장타율을 과대평가
     
     2) 1)의 연장으로 당시 년도 리그의 득점 수준과 구장의 파크팩터* 함께 고려해야 함
           


2. wOBA => 가중 출루율 
           OPS의 단점 1)을 보완한 지표.
   * wOBA = [0.69*(BB-IB) + 0.72*HP + 0.89*HIT + 1.27*H2 + 1.62*H3 + 2.1*HR] / [AB + BB -IB + HP + SF]
   => 출루 이벤트별로 실제 득점가치에 비례한 가중치를 부여해서 출루율을 계산하여 타자가 타석당 생산한 득점을 출루율 범위로 나타낸 것
   
   => 이게 높은 팀일수록 점수 많이 낼 것
   - 단점
   
      1) 고의사구, 실책으로 인한 출루는 고려하지 않음
      
      2) ops의 단점 2)와 동일
      
      3)계산이 복잡하다
      
      
3. GPA = OPS에서 출루율에 가중치를 더 부여한 지표
       = [(1.8*출루율) + 장타율] / 4
       
     - 장점: wOBA는 출루율의 범위로 나오는 반면, GPA는 타율의 범위(약 0.2~0.4)로 나오기 때문에 우리가 더 쉽게 받아들일 수 있다.
           : OPS보다 경기당 평균득점과의 상관계수가 더 높다
           
     - 단점: 출루율에 곱해지는 값이 시대의 득점환경에 따라 달라진다.
      
      
4. wRAA = 리그 평균 대비 득점기여도
        = [(wOBA - 리그 평균 wOBA) / wOBA 범위]*타석 수
        => 0보다 좋으면 리그 평균 이상
      - 단점
         
         1) 누적 값이라 타석 수가 많은 1,2번이 유리
             => 타석 당 득점력 의미하는 wOBA 함께 보는 것이 좋다
             => 잘치는 타자를 앞쪽에 배치하는 것이 최신 야구의 트렌드 이므로 큰 상관 없을 수도 있다.
         2) 구장에 따라 가치가 다르므로 파크팩터를 고려해야 한다.        
         3)계산이 매우 복잡하다


5. wRC+ = 타석 당 생산한 득점을 리그평균, 파크팩터로 조정한 값
        = 타석에서 평균적인 타자보다 얼마나 더 잘했는지 보여주는 지표 (100이 평균)
        
        = [(wRAA/타석 + 리그 득점/타석) + ((1-파크팩터)*리그 득점/타석)] / 리그wRc/타석
        
      - 단점
      
       1) 별 다른 단점은 없지만 계산이 매우 복잡하다  =>  KBreport.com 에서는 존재, 그러나 사용할 수 있을지 의문
       
  
  
* 득점과 관련된 여러 파생변수를 생각해 봤는데, 외부데이터 없이 가지고 있는 데이터로 계산하기에는 2,3이 좋아 보입니다.

### - 어느 팀이 점수를 덜 줄 것인지 어떻게 예측?

1. 평균자책점

   => 가장 직관적인 요소. 그러나 예측해야 함


2. 탈삼진수

   => 삼진은 수비라는 변수가 들어가지 않기 때문에 가장 안전하게 아웃카운트를 늘리는 방법이다.
      삼진 수와 실점이 항상 반비례하지는 않지만 삼진을 많이 잡는 투수는 대개 에이스인 경우가 많다.
   
   => 누적 데이터의 특성 상 던진 이닝수가 많으면 탈삼진 수에 더 유리하기 떄문에 투수의 삼진잡는 능력만 고려하기 위해서는
      삼진/9 로 보정 가능
 
 
3. 장타 허용 = 2루타 허용 + 3루타 허용 + 홈런 허용
   
   => 장타를 허용하게 되면 높은 확률로 점수를 내주기 때문에 이 수치가 낮을 수록 좋은 투수일 것이다.
   
   
4. 9이닝 당 볼넷 수 = (볼넷 수 / 이닝 수) * 9 (이 자료에서는 (볼넷 수 / 이닝 수) * 27)
   
   => 투수에게 제구력 (원하는 곳으로 공을 던지는 능력)은 매우 중요한 능력입니다.
   야구라는 경기의 특성 상 투수에게 매우 유리한 게임이기 때문에 투수가 실수하지 않는다고만 해도 승부가 투수에게 엄청나게 기울 것입니다.
   한경기(9이닝) 당 볼넷을 얼마나 내주는지를 통해 투수의 제구력을 간접적으로 확인하고, 제구력이 좋을 수록 좋은 투수이기 때문에 점수를 덜 내줄 것이다.
   
   
5. 투수 세부지표는 더 찾아보도록 하겠습니다...
